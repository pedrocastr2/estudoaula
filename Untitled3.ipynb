```python
from matplotlib import pyplot as plt
x1 =[1,3,3,4,5,6]
y1=[5,9,7,3,8,2]

x2 =[5,3,3,4,5,6]
y2=[8,0,7,3,7,2]

x3 =[1,2,2,2,3,4]
y3=[1,2,1,2,1,2]
plt.plot(x3,y3,color ='b',linestyle=(0,(3,1,1,0)), label ="linha1")

plt.plot(x1,y1,color ='r', label ="linha2")
plt.bar(x2,y2,color ='g', label ="Barra 1")

plt.title("grafico em barras")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.ylabel("eixo z")
plt.legend()
plt.show()
```
